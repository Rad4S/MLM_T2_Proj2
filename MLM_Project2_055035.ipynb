```python
# Function to calculate descriptive statistics for categorical variables
def rs35_categorical_stats(df, cat_columns):
    rs35_results = []

    for rs35_col in cat_columns:
        rs35_counts = df[rs35_col].value_counts()
        rs35_proportions = rs35_counts / len(df)
        rs35_mode = rs35_counts.idxmax()
        rs35_min = df[rs35_col].min()
        rs35_max = df[rs35_col].max()

        for category, count in rs35_counts.items():
            rs35_results.append({
                "Column": rs35_col,
                "Category": category,
                "Frequency": count,
                "Proportion": rs35_proportions[category],
                "Minimum": rs35_min,
                "Maximum": rs35_max,
                "Mode": rs35_mode
            })

    return pd.DataFrame(rs35_results)

# Function to calculate correlation for categorical columns
def rs35_categorical_correlation(df, cat_columns, method='spearman'):
    rs35_corr_results = pd.DataFrame(index=cat_columns, columns=cat_columns, dtype=float)

    for i in range(len(cat_columns)):
        for j in range(i, len(cat_columns)):
            rs35_col1, rs35_col2 = cat_columns[i], cat_columns[j]
            if i == j:
                # Perfect correlation for the same variable
                corr = 1.0
            else:
                if method == 'spearman':
                    corr, _ = spearmanr(df[rs35_col1].astype('category').cat.codes,
                                        df[rs35_col2].astype('category').cat.codes)
                elif method == 'kendall':
                    corr, _ = kendalltau(df[rs35_col1].astype('category').cat.codes,
                                         df[rs35_col2].astype('category').cat.codes)
            rs35_corr_results.loc[rs35_col1, rs35_col2] = corr
            rs35_corr_results.loc[rs35_col2, rs35_col1] = corr

    return rs35_corr_results

# Retrieve categorical columns from rs35_ds_cat_encoded
rs35_cat_columns = rs35_ds_cat_encoded.columns.tolist()

# Generate descriptive stats and correlations for these columns
rs35_categorical_stats_df = rs35_categorical_stats(rs35_ds_preprocessed, rs35_cat_columns)
rs35_spearman_corr_df = rs35_categorical_correlation(rs35_ds_preprocessed, rs35_cat_columns, method='spearman')
rs35_kendall_corr_df = rs35_categorical_correlation(rs35_ds_preprocessed, rs35_cat_columns, method='kendall')

# Replace NaN correlations with 1 for diagonal elements (if any remain)
rs35_spearman_corr_df.fillna(1, inplace=True)
rs35_kendall_corr_df.fillna(1, inplace=True)

# Display results
print("Descriptive Statistics:")
print(rs35_categorical_stats_df)  # Show all rows

print("\nSpearman Correlation:")
print(rs35_spearman_corr_df)

print("\nKendall Correlation:")
print(rs35_kendall_corr_df)
```

### Non-Categorical Variable

```python
# Descriptive Stats for Noncat Variables

# Function to calculate descriptive statistics for numerical variables
def rs35_numerical_stats(df, num_columns):
    rs35_results = []

    for rs35_col in num_columns:
        rs35_data = df[rs35_col].dropna()  # Exclude NaN values
        rs35_min = rs35_data.min()
        rs35_max = rs35_data.max()
        rs35_mean = rs35_data.mean()
        rs35_median = rs35_data.median()
        rs35_mode = rs35_data.mode().iloc[0] if not rs35_data.mode().empty else np.nan
        rs35_range = rs35_max - rs35_min
        rs35_std_dev = rs35_data.std()
        rs35_skewness = skew(rs35_data)
        rs35_kurt = kurtosis(rs35_data)

        # Coefficient of Variation (CV)
        rs35_cv = rs35_std_dev / rs35_mean if rs35_mean != 0 else np.nan

        # Confidence Interval (95%)
        rs35_conf_interval = norm.interval(0.95, loc=rs35_mean, scale=rs35_std_dev / np.sqrt(len(rs35_data)))

        # Add results
        rs35_results.append({
            "Column": rs35_col,
            "Minimum": rs35_min,
            "Maximum": rs35_max,
            "Mean": rs35_mean,
            "Median": rs35_median,
            "Mode": rs35_mode,
            "Range": rs35_range,
            "Standard Deviation": rs35_std_dev,
            "Skewness": rs35_skewness,
            "Kurtosis": rs35_kurt,
            "Coefficient of Variation": rs35_cv,
            "Confidence Interval (95%)": rs35_conf_interval
        })

    return pd.DataFrame(rs35_results)

# Function to calculate correlation for numerical variables
def rs35_numerical_correlation(df, num_columns, method='pearson'):
    rs35_corr_results = pd.DataFrame(index=num_columns, columns=num_columns, dtype=float)

    for i in range(len(num_columns)):
        for j in range(i, len(num_columns)):
            rs35_col1, rs35_col2 = num_columns[i], num_columns[j]
            if i == j:
                corr = 1.0  # Perfect correlation for the same variable
            else:
                if method == 'pearson':
                    corr, _ = pearsonr(df[rs35_col1].dropna(), df[rs35_col2].dropna())
                elif method == 'spearman':
                    corr, _ = spearmanr(df[rs35_col1].dropna(), df[rs35_col2].dropna())
            rs35_corr_results.loc[rs35_col1, rs35_col2] = corr
            rs35_corr_results.loc[rs35_col2, rs35_col1] = corr

    return rs35_corr_results

# Retrieve numerical columns from rs35_ds_noncat_scaled
rs35_num_columns = rs35_ds_noncat_scaled.columns.tolist()

# Generate statistics and correlations for these columns
rs35_numerical_stats_df = rs35_numerical_stats(rs35_ds_noncat_scaled, rs35_num_columns)
rs35_pearson_corr_df = rs35_numerical_correlation(rs35_ds_noncat_scaled, rs35_num_columns, method='pearson')
rs35_spearman_corr_df = rs35_numerical_correlation(rs35_ds_noncat_scaled, rs35_num_columns, method='spearman')

# Replace NaN correlations with 1 for diagonal elements (if any remain)
rs35_pearson_corr_df.fillna(1, inplace=True)
rs35_spearman_corr_df.fillna(1, inplace=True)

# Display results
print("Numerical Statistics:")
print(rs35_numerical_stats_df)

print("\nPearson Correlation:")
print(rs35_pearson_corr_df)

print("\nSpearman Correlation:")
print(rs35_spearman_corr_df)
```

```python
from sklearn.model_selection import KFold, train_test_split
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np
import requests
from io import StringIO
from scipy import stats
import time
import psutil
import os
import matplotlib.pyplot as plt
import seaborn as sns


# Load the dataset (assuming rs35_ds is already loaded)
# The code to load rs35_ds is given in the notebook as shown in the context

# Replace with your actual file ID
file_id = '1Fh5cH34r8ikd_69YRp3zPEv7kl1xWwRH'

# Construct the URL
url = f'https://drive.google.com/uc?id={file_id}'

# Fetch the data using requests
response = requests.get(url)
response.raise_for_status()  # Raise an exception for bad responses

# Read the data into a pandas DataFrame using StringIO
rs35_df = pd.read_csv(StringIO(response.text))
rs35_ds = rs35_df.sample(n = 5001, random_state=55035)

# --- Encoding Categorical Variables
# Define mappings
import_export_mapping = {'Export': 0, 'Import': 1}
category_mapping = {'Clothing': 0, 'Electronics': 1, 'Furniture': 2, 'Machinery': 3, 'Toys': 4}
shipping_mapping = {'Air': 0, 'Land': 1, 'Sea': 2}
payment_mapping = {'Cash on Delivery': 0, 'Net 30': 1, 'Net 60': 2, 'Prepaid': 3}

# Apply mappings
rs35_ds['Import_Export'] = rs35_ds['Import_Export'].map(import_export_mapping)
rs35_ds['Category'] = rs35_ds['Category'].map(category_mapping)
rs35_ds['Shipping_Method'] = rs35_ds['Shipping_Method'].map(shipping_mapping)
rs35_ds['Payment_Terms'] = rs35_ds['Payment_Terms'].map(payment_mapping)

# Define features and target variable
X = rs35_ds[['Quantity', 'Value', 'Weight', 'Import_Export', 'Category', 'Shipping_Method', 'Payment_Terms']]
y = rs35_ds['Category']

# Perform 80-20 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=55035)

# Set up K-Fold Cross-Validation on the training set
kf = KFold(n_splits=5, shuffle=True, random_state=55035)

# Initialize models
logistic_model = LogisticRegression(solver='liblinear', random_state=55035)
decision_tree_model = DecisionTreeClassifier(random_state=55035)
random_forest_model = RandomForestClassifier(random_state=55035)

models = {
    "Logistic Regression": logistic_model,
    "Decision Tree": decision_tree_model,
    "Random Forest": random_forest_model
}

# Dictionary to store results
results = {}


def memory_usage_mb():
    process = psutil.Process(os.getpid())
    mem_bytes = process.memory_info().rss
    mem_mb = mem_bytes / (1024 * 1024)
    return mem_mb

# Perform K-Fold Cross-Validation for each model
for model_name, model in models.items():
    fold_accuracies = []
    print(f"Model: {model_name}")

    # Time Tracking
    start_train = time.time()
    mem_before_train = memory_usage_mb()

    for fold, (train_index, val_index) in enumerate(kf.split(X_train, y_train)):
        print(f"  Fold {fold + 1}")
        X_train_fold, X_val_fold = X_train.iloc[train_index], X_train.iloc[val_index]
        y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]

        # Train the model
        model.fit(X_train_fold, y_train_fold)

        # Make predictions on the validation set
        y_pred = model.predict(X_val_fold)

        # Evaluate the model
        accuracy = accuracy_score(y_val_fold, y_pred)
        print(f"    Accuracy: {accuracy:.4f}")
        fold_accuracies.append(accuracy)

    # Memory Usage
    mem_after_train = memory_usage_mb()

    # Prediction Time
    start_predict = time.time()
    y_pred_test = model.predict(X_test) # Predict on the test set to get proper prediction time
    end_predict = time.time()
    predict_time = end_predict-start_predict

    # End training timer
    end_train = time.time()
    train_time = end_train - start_train

    # Calculate average accuracy across folds
    mean_accuracy = np.mean(fold_accuracies)
    print(f"  Mean Accuracy across all folds: {mean_accuracy:.4f}")
    print(f"  Training time: {train_time:.4f} seconds")
    print(f"  Memory usage before training: {mem_before_train:.2f} MB")
    print(f"  Memory usage after training: {mem_after_train:.2f} MB")
    print(f"  Prediction time : {predict_time:.4f} seconds\n")
    results[model_name] = mean_accuracy
    results[model_name + "_fold_accuracies"] = fold_accuracies
    results[model_name + "_train_time"] = train_time
    results[model_name + "_memory_usage"] = mem_after_train - mem_before_train
    results[model_name + "_predict_time"] = predict_time

    if model_name == "Random Forest":
        rf_model = RandomForestClassifier(random_state=55035)
        rf_model.fit(X_train,y_train)
        importances = rf_model.feature_importances_
        feature_names = X_train.columns

        feature_importance = pd.DataFrame({'Feature':feature_names,'Importance':importances})
        feature_importance = feature_importance.sort_values('Importance',ascending=False)
        print("Random Forest Feature Importance:")
        print(feature_importance)
        # Plotting feature importances
        plt.figure(figsize=(10, 6))
        sns.barplot(x='Importance', y='Feature', data=feature_importance)
        plt.title('Random Forest Feature Importances')
        plt.show()
    if model_name == "Logistic Regression":
       logistic_model = LogisticRegression(solver='liblinear',random_state=55035)
       logistic_model.fit(X_train,y_train)
       coefficients = logistic_model.coef_
       feature_names = X_train.columns
       # Calculate absolute coefficient values
       abs_coefs = np.abs(coefficients)

       # Average coefficients if multiclass
       if len(abs_coefs)>1:
        abs_coefs = np.mean(abs_coefs, axis=0)
       else:
           abs_coefs = abs_coefs[0]

       feature_importance = pd.DataFrame({'Feature':feature_names, 'Importance':abs_coefs})
       feature_importance = feature_importance.sort_values('Importance',ascending=False)
       print("Logistic Regression Feature Importance:")
       print(feature_importance)
       # Plotting feature importances
       plt.figure(figsize=(10, 6))
       sns.barplot(x='Importance', y='Feature', data=feature_importance)
       plt.title('Logistic Regression Feature Importances')
       plt.show()

    if model_name == "Decision Tree":
        dt_model = DecisionTreeClassifier(random_state=55035)
        dt_model.fit(X_train, y_train)
        importances = dt_model.feature_importances_
        feature_names = X_train.columns

        feature_importance = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
        feature_importance = feature_importance.sort_values('Importance', ascending=False)
        print("Decision Tree Feature Importance:")
        print(feature_importance)
           # Plotting feature importances
        plt.figure(figsize=(10, 6))
        sns.barplot(x='Importance', y='Feature', data=feature_importance)
        plt.title('Decision Tree Feature Importances')
        plt.show()

# Determine the best model based on mean accuracy
best_model = max(results, key=lambda k: results[k] if "_fold_accuracies" not in k and "_train_time" not in k and "_memory_usage" not in k and "_predict_time" not in k else -np.inf)
best_accuracy = results[best_model]
print(f"The best model based on mean accuracy is: {best_model} with a mean accuracy of {best_accuracy:.4f}\n")

# --- Hypothesis Testing ---
# Hypothesis: Random Forest is better than Logistic Regression.
# H0: There is no difference in the performance of both models
# H1: Random Forest is better than Logistic Regression.
alpha = 0.05  # Significance level
if "Random Forest_fold_accuracies" in results and "Logistic Regression_fold_accuracies" in results:
    rf_accuracies = results["Random Forest_fold_accuracies"]
    lr_accuracies = results["Logistic Regression_fold_accuracies"]

    # Perform a paired t-test (since we are comparing model results on same folds)
    t_stat, p_value = stats.ttest_rel(rf_accuracies, lr_accuracies, alternative='greater')

    print("Hypothesis Test Results:")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis. Random Forest is significantly better than Logistic Regression.")
    else:
        print("Conclusion: Fail to reject the null hypothesis. There is not enough evidence to suggest that Random Forest is significantly better than Logistic Regression.")
else:
    print("Hypothesis Test not possible as Random Forest and/or Logistic Regression accuracies are missing.")

# Model Run Statistics
print("\nModel Run Statistics:")
for model_name in models.keys():
    print(f"  {model_name}:")
    print(f"    Training Time: {results[model_name + '_train_time']:.4f} seconds")
    print(f"    Memory Usage: {results[model_name + '_memory_usage']:.2f} MB")
    print(f"    Prediction Time: {results[model_name + '_predict_time']:.4f} seconds")
```
